```python
# Meant to move a servo motor with second camera mounted on top, if first camera detects object
import serial
import time
import threading
from queue import Queue

# Initialize pySerial connection
arduino = serial.Serial('COM3', 9600, timeout=1)
time.sleep(2)

def control_led(led_number, state):
    command = f"{led_number}{'1' if state == 'on' else '0'}"
    arduino.write(command.encode())
    time.sleep(0.1)
    response = arduino.readline().decode().strip()
    print(f"Arduino says: {response}")


def move_motor(angle):
    motor_command = f"MOTOR:{angle}"
    arduino.write(motor_command.encode())
    time.sleep(0.1)
    response = arduino.readline().decode().strip()
    print(f"Motor Response: {response}")

def process_frame():
    while True:
        if not frame_queue.empty():
            frame = frame_queue.get()
            results = model(frame)
            # Process results

frame_queue = Queue()
thread = threading.Thread(target=process_frame, daemon=True)
thread.start()
   

# Initialize webcam or video feed
the_url = "rtsp://admin:Password@192.168.1.108:554/cam/realmonitor?channel=1&subtype=1"
cap = cv2.VideoCapture(the_url)
cap.set(cv2.CAP_PROP_BUFFERSIZE, 1)  # Reduce buffer size
cap.set(cv2.CAP_PROP_FPS, 30)        # Set FPS (frames per second)

if not cap.isOpened():
    print("Error: Could not open video stream from Main camera.")
else:
    print("Connected to the MAIN camera.")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("No frame captured.")
        break

    # Run detection
    results = model(frame)

    # Filter detected objects by confidence
    filtered_objects = []
    
    if hasattr(results, 'boxes') and results.boxes is not None:
       for cls, conf in zip(results.boxes.cls, results.boxes.conf):
           if conf > 0.5:  # Confidence filter
               print(f"Detected: {results.names[int(cls)]} with confidence {conf:.2f}")


    # Detection conditions
    person_detected = any(obj == 'person' for obj, _ in filtered_objects)
    threat_detected = any(
        obj in ['gun', 'simulation gun', 'face covering'] for obj, _ in filtered_objects if person_detected
    )
    high_confidence = any(conf > 0.7 for _, conf in filtered_objects if threat_detected)

    # Person detection: Light LED1
    control_led(1, 'on' if person_detected else 'off')

    # Threat detection: Light LED2, send GSM message, move motor
    if threat_detected:
        control_led(2, 'on')
        print("Person detected. Sending commands to Arduino.")
        commands = ["ON\n", "OFF\n"]  # Example commands for Arduino

        for command in commands:
            print(f"Sending: {command.strip()}")
            arduino.write(command.encode())  # Send command to Arduino
            time.sleep(1)  # Wait for Arduino to process
    else:
        control_led(2, 'off')

    # High confidence detection: Light LED3, send GSM message
    if high_confidence:
        control_led(3, 'on')
        move_motor(90)
        print("Person detected. Sending commands to Arduino.")
        commands = ["ON\n", "OFF\n"]  # Example commands for Arduino

        for command in commands:
            print(f"Sending: {command.strip()}")
            arduino.write(command.encode())  # Send command to Arduino
            time.sleep(1)  # Wait for Arduino to process
    else:
        control_led(3, 'off')

    # Display the annotated frame
    frame_queue.put(frame)
    cv2.imshow('Object Detection', frame)

    # Stop on key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
